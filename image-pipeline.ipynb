```python
from transformers import LayoutLMv2Processor, LayoutLMv2ForTokenClassification

model_name = "microsoft/layoutlmv2-large-uncased"
processor = LayoutLMv2Processor.from_pretrained(model_name)
model = LayoutLMv2ForTokenClassification.from_pretrained(model_name)
```

```python
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
from PIL import Image

# Lade das Bild
image_path = "./data/table.png"
image = Image.open(image_path).convert("RGB")

# Verarbeite das Bild
encoding = processor(image, return_tensors="pt")
input_ids = encoding["input_ids"].to(device)
attention_mask = encoding["attention_mask"].to(device)
token_type_ids = encoding["token_type_ids"].to(device)
bbox = encoding["bbox"].to(device)
image = encoding["image"].to(device)
```

```python
with torch.no_grad():
    outputs = model(input_ids=input_ids, bbox=bbox, image=image, attention_mask=attention_mask, token_type_ids=token_type_ids)

logits = outputs.logits
predictions = torch.argmax(logits, dim=-1)
predictions
```

```python
label_map = {0: 'O', 1: 'B-KEY', 2: 'I-KEY', 3: 'B-VALUE', 4: 'I-VALUE'}
predicted_labels = [label_map[pred.item()] for pred in predictions[0]]
predicted_labels
```

```python
tokens = processor.tokenizer.convert_ids_to_tokens(input_ids[0])
tokens
```

```python
keys = []
values = []
current_key = None

for token, label in zip(tokens, predicted_labels):
    if label == 'B-KEY':
        if current_key:
            keys.append(current_key)
        current_key = token
    elif label == 'I-KEY' and current_key:
        current_key += f" {token}"
    elif label == 'B-VALUE':
        if current_key:
            values.append((current_key, token))
            current_key = None

print(keys)
print(values)
```
